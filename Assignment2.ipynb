```python
#numpy--------------------------------task1----------------------------------

import numpy as np

## 1. Creating Arrays ##

# 1D array with integers 1-10
array_1d = np.arange(1, 11)

# 2D array (3x3) with integers 1-9
array_2d = np.arange(1, 10).reshape(3, 3)

# 3D array (3x5x3) with random floats between 0 and 1
array_3d = np.random.rand(3, 5, 3)

## 2. Displaying Array Properties ##

def display(arr, name):
    print(f"\n{name} Properties:")
    print("Array:")
    print(arr)
    print(f"Shape: {arr.shape}")
    print(f"Size: {arr.size}")
    print(f"Data type: {arr.dtype}")

# Display properties for each array
display(array_1d, "1D Array")
display(array_2d, "2D Array")
display(array_3d, "3D Array")
```

```python
# TASK2---------------------------------------------------------

# 1. Create array from list
data = [10, 20, 30, 40, 50, 60, 70, 80, 90]
arr = np.array(data)

# 2. Perform operations
first_three = arr[:3]  # First three elements
alternate = arr[::2]  # Every alternate element
reversed_arr = arr[::-1]  # Reversed array

# Display results

print("\nOriginal array:", arr)
print("First three elements:", first_three)
print("Every alternate element:", alternate)
print("Reversed array:", reversed_arr)
```

```python
# TASK3------------------------------------------
import numpy as np

# Step 1: Create two NumPy arrays with 5 random integers between 1 and 20
A = np.random.randint(1, 21, size=5)
B = np.random.randint(1, 21, size=5)

print("Array A:", A)
print("Array B:", B)

# Step 2a: Element-wise operations
print("\nElement-wise Addition:      ", A + B)
print("Element-wise Subtraction:   ", A - B)
print("Element-wise Multiplication:", A * B)
print("Element-wise Division:      ", A / B)

# Step 2b: Dot product
print("\nDot Product of A and B:     ", np.dot(A, B))

# Step 2c: Statistics on array A
print("\nMean of A:                  ", np.mean(A))
print("Median of A:                ", np.median(A))
print("Standard Deviation of A:    ", np.std(A))
print("Variance of A:              ", np.var(A))

# Step 2d: Max/Min of array B and their indices
print("\nMaximum of B:               ", np.max(B))
print("Index of Maximum in B:      ", np.argmax(B))
print("Minimum of B:               ", np.min(B))
print("Index of Minimum in B:      ", np.argmin(B))
```

```python
# TASK4------------------------------------------------------------------------
import numpy as np

# Step 1: Create a 1D array with integers from 1 to 12
arr = np.arange(1, 13)
print("Original 1D Array:")
print(arr)


arr2d = arr.reshape(4, 3)
print("\nReshaped to 2D (4, 3):")
print(arr2d)

# Step 2b: Reshape into 3D array of shape (2, 2, 3)
arr3d = arr.reshape(2, 2, 3)
print("\nReshaped to 3D (2, 2, 3):")
print(arr3d)

# Step 2c: Transpose the 2D array and display its shape
arr2dt = arr2d.T
print("\nTransposed 2D Array:")
print(arr2dt)
print("Shape after transpose:", arr2dt.shape)
```

```python
# TASK 5--------------------------------------------------

import numpy as np

# Step 1: Create array with 15 random numbers from 10 to 50
a = np.random.randint(10, 51, size=15)
print("Original array:", a)

# Step 2a: Elements greater than 25
gt25 = a[a > 25]
print("Greater than 25:", gt25)

# Step 2b: Replace elements less than 30 with 0
a_mod = a.copy()
a_mod[a_mod < 30] = 0
print("After replacing <30 with 0:", a_mod)

# Step 2c: Count elements divisible by 5
count5 = np.sum(a % 5 == 0)
print("Count divisible by 5:", count5)
```

```python
# TASK7-----------------------------------------------------
import numpy as np
import time

# Step 1: Generate two large random arrays of size (100, 100)
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)

# Step 2: Measure time before multiplication
start = time.time()

# Step 3: Perform matrix multiplication
C = np.dot(A, B)

# Step 4: Try to compute determinant and inverse (if possible)
try:
    det = np.linalg.det(C)
    inv = np.linalg.inv(C)
    det_status = "Determinant computed"
    inv_status = "Inverse computed"

    
except np.linalg.LinAlgError:
    det = None
    inv = None
    det_status = "Determinant not computable"
    inv_status = "Matrix is singular; inverse not possible"

# Step 5: Measure time after all operations
end = time.time()
elapsed = end - start

# Step 6: Output results
print("Matrix multiplication completed.")
print(det_status)
print(inv_status)
print("Time taken: {:.4f} seconds".format(elapsed))

# Optional: print determinant value (rounded)
if det is not None:
    print("Determinant: {:.4f}".format(det))
```
